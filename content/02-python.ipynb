```python
dic={}
for i in range(1,21):
  dic.update({i:i*i})
```

```python
dic={i:i**2 for i in range(1,21)}
```

```python
dic
```

```python
ls1=[i for i in range(1,21) if i%2==0]
ls2=[i for i in range(1,21) if i%2!=0]
```

```python
dic={}
for i,j in zip(ls1,ls2):
  dic.update({i:j})
```

```python
dic={i:j for i,j in zip(ls1,ls2)}
```
